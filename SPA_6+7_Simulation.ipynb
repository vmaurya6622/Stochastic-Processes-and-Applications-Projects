```python
# Question 06
'''
Simulate the random variables X1, X2, . . . , Xn for five different values of n, to
show that the corresponding CDFs of Zn demonstrate the property illustrated in Figure 1.9 of
the book. Repeat the above for when the random variables are normal and lognormal. Let the
normal and the lognormal random variables have the same mean and variance. See https://en.
wikipedia.org/wiki/Log-normal_distribution for setting its parameters to obtain a desired
mean and variance.
Based on your simulations, comment on the convergence of the CDFs to a unit step for the
two distributions.'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate lognormal parameters
def Lognormal_Calculation(mu, sigma):
    Ln_Sigma = np.sqrt(np.log((sigma / mu)**2 + 1))
    Ln_Mean = np.log(mu) - Ln_Sigma**2 / 2
    return Ln_Mean, Ln_Sigma

# Parameters
mu, sigma = 1, 1  # Mean and standard deviation for normal distribution
Ln_Mean, Ln_Sigma = Lognormal_Calculation(mu, sigma)

# Sample Sample_sizes
Sample_Sizes = [5, 20, 30, 50, 80]  # Sample sizes
simulations = 150           # Number of simulations
colors = ['b', 'g', 'r', 'c', 'm']  # Colors

### Plot for Normal Distribution ###
plt.figure(figsize=(10, 6))
for idx, n in enumerate(Sample_Sizes):
    # Generating  normal distribution data
    data = np.random.normal(mu, sigma, (simulations, n))
    averages = data.mean(axis=1)

    # Calculate the empirical CDF
    values, base = np.histogram(averages, bins=1000, density=True)
    cumulative = np.cumsum(values) * np.diff(base[:2])

    # Plot the step function for CDF
    plt.step(base[:-1], cumulative, where='mid', color=colors[idx], label=f'n = {n}')

### Plot for Normal Distribution ###
def normal_distribution_plotting():
    plt.xlim(-.5, 2.5)  # Set x-axis limits for normal distribution
    plt.ylim(0, 1)       # CDF always ranges from 0 to 1
    plt.xlabel(r'$Z_n$')
    plt.ylabel('CDF')
    plt.title(r'Empirical CDF of $Z_n$ for Normal distribution')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

### Plot for Lognormal Distribution ###
plt.figure(figsize=(10, 6))
for idx, n in enumerate(Sample_Sizes):
    # Generate lognormal distribution data
    data = np.random.lognormal(Ln_Mean, Ln_Sigma, (simulations, n))
    averages = data.mean(axis=1)

    # Calculate the empirical CDF
    values, base = np.histogram(averages, bins=1000, density=True)
    cumulative = np.cumsum(values) * np.diff(base[:2])

    # Plot the step function for CDF
    plt.step(base[:-1], cumulative, where='mid', color=colors[idx], label=f'n = {n}')

# Customize the plot for Lognormal Distribution
def lognormal_distribution_plotting():
    plt.xlim(0, 3)  # Set x-axis limits for lognormal distribution (strictly positive)
    plt.ylim(0, 1)   # CDF always ranges from 0 to 1
    plt.xlabel(r'$Z_n$')
    plt.ylabel('CDF')
    plt.title(r'Empirical CDF of $Z_n$ for Lognormal distribution')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

normal_distribution_plotting
lognormal_distribution_plotting
```


### Observations:
1. **Empirical CDF of \( Z_n \) for Normal Distribution:**
   - As the sample size \( n \) increases, the empirical CDF becomes steeper and more defined. For smaller sample sizes (e.g., \( n = 5 \)), the CDF appears smoother and less steep, while for larger \( n \) values (e.g., \( n = 80 \)), it approaches a step-like pattern.
   - In the case of the normal distribution, the CDFs converge around the mean (approximately 0), and this concentration around the mean strengthens as the sample size increases.

2. **Empirical CDF of \( Z_n \) for Lognormal Distribution:**
   - For the lognormal distribution, the CDF is skewed, which aligns with the distribution's inherent skewness.
   - Similar to the normal distribution, as the sample size increases, the CDF becomes steeper and sharper, approaching a step-like shape. However, the values remain positive and extend further to the right, reflecting the skewness of the distribution.

### Conclusion:
- **Normal Distribution:** As sample size grows, the empirical CDF for the normal distribution increasingly mirrors the true CDF, in line with the central limit theorem. The distribution centers more around the mean (around 0) as \( n \) increases.

- **Lognormal Distribution:** The empirical CDF of the lognormal distribution retains its right skew, but with larger sample sizes, it converges more tightly. Unlike the normal distribution, the lognormal CDF remains skewed to the right and lacks symmetry.

**Summary:** As the sample size increases, the accuracy of the empirical CDF improves for both the normal and lognormal distributions, with each reflecting its respective characteristics more clearly.


```python
# Question 07
'''
Simulate a stochastic process Xt; t  1, where the random variables Xt are
IID Bernoulli(0.8). Specifically, generate an ensemble of 1000 sample functions, where each sample
function is generated for t = 1, 2, . . . , 2000. For every t, calculate the empirical mean (based on the
1000 sample functions) of the random variable X(t). Plot the histogram of the empirical means.
Now for each sample function (a total of 2000) calculate the sample mean (time-average) of the
sample function. Plot a histogram of the sample means. Compare the two histograms. Describe
what you see and explain why what you see is what is expected.'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.8  # Probability of success in the Bernoulli distribution
p2 = 0.5  # Probability of success in the Bernoulli distribution
num_samples = 1000  # Number of sample functions
T = 2000  # Length of each sample function

# Step 1: Generating the sample functions
data = np.random.binomial(1, p, (num_samples, T))
data2 = np.random.binomial(1, p2, (num_samples, T))

# Step 2: Calculating empirical means at each time t
empirical_means = data.mean(axis=0)

# Step 3: Calculating sample means for each sample function
sample_means = data.mean(axis=1)

# Step 4: Plotting histograms
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, edgecolor="black", color="violet", alpha=0.7)
plt.title("Histogram of Sample Means for Each Sample Function (P = 0.8)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.subplot(1, 2, 1)
plt.hist(empirical_means, edgecolor="black", bins=30, color="pink", alpha=0.7)
plt.title("Histogram of Empirical Means Across Time Steps (P = 0.8)")
plt.xlabel("Empirical Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()





# Step 2: Calculating empirical means at each time t
empirical_means2 = data2.mean(axis=0)

# Step 3: Calculating sample means for each sample function
sample_means2 = data2.mean(axis=1)

# Step 4: Plotting histograms
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 2)
plt.hist(sample_means2, bins=30, edgecolor="black", color="violet", alpha=0.7)
plt.title("Histogram of Sample Means for Each Sample Function (P = 0.5)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.subplot(1, 2, 1)
plt.hist(empirical_means2, edgecolor="black", bins=30, color="pink", alpha=0.7)
plt.title("Histogram of Empirical Means Across Time Steps (P = 0.5)")
plt.xlabel("Empirical Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```


**Observations:**

**Comparison Between Bernoulli(0.8) and Bernoulli(0.5):**
- **Shape of the Distribution:**
  * For both Bernoulli(0.8) and Bernoulli(0.5), the empirical means show more variability compared to the time-averaged means. This occurs because empirical means are based on fewer observations at any given time step.
  * The time-averaged means cluster tightly around the expected values (0.8 for Bernoulli(0.8) and 0.5 for Bernoulli(0.5)), highlighting the Law of Large Numbers at work.

- **Spread of the Distribution:**
  * The empirical means for Bernoulli(0.5) exhibit a slightly wider spread compared to Bernoulli(0.8), as the 0.5 probability of success results in more balanced outcomes between 0s and 1s, increasing variability.
  * The time-averaged means for both distributions are closely centered around the expected values, showing clear convergence.

**Conclusion:**
- Both empirical and time-averaged means align with the expectations of the Law of Large Numbers. Over time, the sample means for both Bernoulli distributions converge toward their respective expected values.
- The empirical means of Bernoulli(0.5) show more variability compared to Bernoulli(0.8), but both distributions exhibit similar convergence behavior in their time-averaged means.
